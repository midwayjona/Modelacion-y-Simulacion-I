## Función para Calcualar la Distancia

```python
def distancia_entre_puntos(punto1, punto2):
    x1, y1 = punto1
    x2, y2 = punto2
    distancia = np.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return distancia
```

```python
def generate_random_solution(arr, initial_val):
    arr.remove(initial_val)
    random.shuffle(arr)
    solution = [initial_val] + arr + [initial_val]
    ruta = [[solution[i], solution[i+1]] for i in range(len(solution) - 1)]

    return solution, ruta
```

## Población Inicial 

```python
def generar_poblacion_inicial(k=200):
    poblacion_0 = []
    for i in range(0, k):
        temp = list(ciudades.keys())
        solucion_aleatoria, _ = generate_random_solution(temp, 'C1')
        poblacion_0.append(solucion_aleatoria)
    
    return poblacion_0
```

```python
def calcular_fitness(solucion, df_distancias):
    distancia_total = 0
    for index, ciudad in enumerate(solucion):
        try:
            ciudad_inicial = solucion[index]
            ciudad_siguiente = solucion[index + 1]
            distancia = df_distancias.loc[ciudad_inicial, ciudad_siguiente]
            #print(f"{ciudad_inicial} -> {ciudad_siguiente} = {distancia}")
            distancia_total = distancia_total + distancia
        except:
            pass

    return distancia_total
```

```python
def calcular_fitness_df(poblacion):
    fitness_solucion_dict = {}
    for individuo in poblacion:
        fitness_solucion_dict[''.join(individuo)] = calcular_fitness(individuo, matriz_distancias)

    df_poblacion_fitnesses = pd.DataFrame.from_dict(fitness_solucion_dict, orient='index').reset_index()
    df_poblacion_fitnesses.columns = ['individuo', 'fitness']
    return df_poblacion_fitnesses
```

```python
poblacion_0 = generar_poblacion_inicial()
df_finess_poblacion = calcular_fitness_df(poblacion_0)
df_finess_poblacion
```

## CrossOver - Combinación de Individuos

```python
def gen_parejas(poblacion):
    soluciones = poblacion['individuo']
    parejas = []
    for _ in range(len(poblacion)):
        pareja = soluciones.sample(2)
        parejas.append(pareja.tolist())

    return parejas
```

```python
parejas = gen_parejas(mejores_poblacion)
```

```python
def robust_crossover(parent1, parent2, n = 5):
    out = []

    parent1 = parent1.split('C')[2: -1]
    parent2 = parent2.split('C')[2: -1]
    
    infoParent1 = parent1[-n:]
    child1 = [elemento for elemento in parent2 if elemento not in infoParent1] + infoParent1
    infoParent2 = parent2[-n:]
    child2 = [elemento for elemento in parent1 if elemento not in infoParent2] + infoParent2

    child1.insert(0, 1)
    child1.insert(len(child1) + 1, 1)
    child2.insert(0, 1)
    child2.insert(len(child2) + 1, 1)

    child1 = ['C'+str(val) for val in child1]
    child2 = ['C'+str(val) for val in child2]

    out.append(child1)
    out.append(child2)

    return out    
```

```python
def crossover(parejas):
    new_generation = []
    for pareja in parejas:
        childs = robust_crossover(pareja[0], pareja[1])
        new_generation.extend(childs)
    return new_generation
```

```python
nueva_generacion = crossover(parejas)
calcular_fitness_df(nueva_generacion).nsmallest(100, 'fitness')
```

```python
def mutacion(generation, p_m=0.1):
    N = list(range(len(generation)))
    cantidad_mutaciones = int(len(N)*p_m)
    elmentos_a_mutar = random.sample(N, cantidad_mutaciones)

    for a_mutar in elmentos_a_mutar:
        individuo = generation[a_mutar]
        individuo = [i for i in individuo if i != 'C1']
        n = len(individuo)
        swap_index = random.sample(list(range(n)), 2)
        tempInd = individuo.copy()
        individuo[swap_index[0]] = individuo[swap_index[1]]
        individuo[swap_index[1]] =  tempInd[swap_index[0]]
        generation[a_mutar] = ['C1'] + individuo + ['C1']

    return generation    
```

## Inicialización del Problema

```python
def init_problem(cantidad_ciudades):

    x = np.round(np.random.rand(cantidad_ciudades)*10, 4)
    y = np.round(np.random.rand(cantidad_ciudades)*10, 4)

    ciudades = dict()

    for i, (xi, yi) in enumerate(zip(x, y)):
        ciudades[f'C{i+1}'] = (xi, yi)

    matriz_distancias = pd.DataFrame(index=ciudades.keys(), columns=ciudades.keys())
    matriz_distancias.fillna(0.0, inplace=True)
    
    for par in pares:
        puntoA = ciudades[par[0]]
        puntoB = ciudades[par[1]]
        matriz_distancias.loc[par[0], par[1]] = distancia_entre_puntos(puntoA, puntoB)

    return matriz_distancias
```

# Simulación

```python
generaciones = 500
n_ciudades = 10
porcentaje_seleccion = 0.5
n_pinicial = 200

matriz_distancias = init_problem(n_ciudades)
poblacion_inicial = generar_poblacion_inicial(n_pinicial)
df_fitness_poblacion = calcular_fitness_df(poblacion_inicial)
n_mejores = int(len(poblacion_inicial)*porcentaje_seleccion)

mean_fitness = []
min_fitness = []

for gen in range(generaciones):

    # seleccion natural
    mejores_de_generacion = df_fitness_poblacion.nsmallest(n_mejores, 'fitness')

    #crossover
    parejas = gen_parejas(mejores_de_generacion)
    new_gen_crossover = crossover(parejas)

    # mutacion
    mutated_generation = mutacion(new_gen_crossover, p_m=0.1)
    
    # tabla de fitness de la generacion 
    df_fitness_poblacion = calcular_fitness_df(mutated_generation)

    mean_fitness.append(df_fitness_poblacion['fitness'].mean())
    min_fitness.append(df_fitness_poblacion['fitness'].min())
```

```python
plt.plot(mean_fitness)
fitness_promedio = round(sum(mean_fitness)/len(mean_fitness), 4)
plt.title(f"Fitness Promedio - {fitness_promedio}")
plt.show()
```
